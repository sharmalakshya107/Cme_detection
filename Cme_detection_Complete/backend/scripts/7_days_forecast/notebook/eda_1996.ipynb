## Correlation Analysis

```python
# Select numeric columns (exclude Datetime, Year, DOY, Hour)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
exclude_cols = ['Year', 'DOY', 'Hour']
analysis_cols = [col for col in numeric_cols if col not in exclude_cols]

print(f"Analyzing correlations for {len(analysis_cols)} parameters")
print(analysis_cols)
```

```python
# Calculate correlation matrix
correlation_matrix = df[analysis_cols].corr()

# Create correlation heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm', 
            center=0,
            square=True,
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})

plt.title('Correlation Heatmap ', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# Find highly correlated pairs (|correlation| > 0.7)
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr_pairs.append({
                'Feature_1': correlation_matrix.columns[i],
                'Feature_2': correlation_matrix.columns[j],
                'Correlation': correlation_matrix.iloc[i, j]
            })

high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('Correlation', ascending=False)
print("\nHighly Correlated Feature Pairs (|r| > 0.7):")
print(high_corr_df)
```

```python
# Outlier Detection using IQR Method
print("="*70)
print("OUTLIER DETECTION (IQR Method)")
print("="*70)

outlier_summary = []

for col in analysis_cols:
    data_clean = df[col].dropna()
    
    if len(data_clean) > 0:
        Q1 = data_clean.quantile(0.25)
        Q3 = data_clean.quantile(0.75)
        IQR = Q3 - Q1
        
        # Define outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outliers = ((data_clean < lower_bound) | (data_clean > upper_bound)).sum()
        outlier_pct = (outliers / len(data_clean)) * 100
        
        outlier_summary.append({
            'Parameter': col,
            'Count': len(data_clean),
            'Q1': Q1,
            'Median': data_clean.median(),
            'Q3': Q3,
            'IQR': IQR,
            'Lower_Bound': lower_bound,
            'Upper_Bound': upper_bound,
            'Outliers': outliers,
            'Outlier_Percent': outlier_pct
        })

outlier_df = pd.DataFrame(outlier_summary).sort_values('Outlier_Percent', ascending=False)
print("\nOutlier Summary (sorted by outlier percentage):")
print(outlier_df[['Parameter', 'Outliers', 'Outlier_Percent', 'Lower_Bound', 'Upper_Bound']].to_string(index=False))

print("\n" + "="*70)
```

```python
# Visualize outliers by parameter
fig, axes = plt.subplots(6, 3, figsize=(20, 24))
axes = axes.flatten()

for idx, col in enumerate(analysis_cols):
    data_clean = df[col].dropna()
    
    if len(data_clean) > 0:
        Q1 = data_clean.quantile(0.25)
        Q3 = data_clean.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identify outliers
        outliers_mask = (data_clean < lower_bound) | (data_clean > upper_bound)
        outliers = data_clean[outliers_mask]
        normal = data_clean[~outliers_mask]
        
        # Create scatter plot
        axes[idx].scatter(range(len(normal)), normal, alpha=0.3, s=1, color='blue', label='Normal')
        if len(outliers) > 0:
            outliers_indices = [i for i, val in enumerate(data_clean) if val in outliers.values]
            axes[idx].scatter(outliers_indices, outliers, alpha=0.6, s=3, color='red', label='Outliers')
        
        axes[idx].axhline(y=lower_bound, color='green', linestyle='--', linewidth=1, alpha=0.7, label='Bounds')
        axes[idx].axhline(y=upper_bound, color='green', linestyle='--', linewidth=1, alpha=0.7)
        axes[idx].set_title(f'{col}\n({len(outliers)} outliers, {len(outliers)/len(data_clean)*100:.1f}%)', fontsize=9)
        axes[idx].set_ylabel('Value', fontsize=8)
        axes[idx].grid(True, alpha=0.3)
        if idx == 0:
            axes[idx].legend(fontsize=7)

# Hide unused subplots
for idx in range(len(analysis_cols), len(axes)):
    axes[idx].axis('off')

plt.suptitle('Outlier Detection - All Parameters', fontsize=16, y=0.995)
plt.tight_layout()
plt.show()
```

```python
# Statistical tests for outliers - Z-score method
print("="*70)
print("OUTLIER DETECTION (Z-Score Method, |z| > 3)")
print("="*70)

zscore_summary = []

for col in analysis_cols:
    data_clean = df[col].dropna()
    
    if len(data_clean) > 0:
        mean = data_clean.mean()
        std = data_clean.std()
        
        if std > 0:
            z_scores = np.abs((data_clean - mean) / std)
            outliers = (z_scores > 3).sum()
            outlier_pct = (outliers / len(data_clean)) * 100
            
            zscore_summary.append({
                'Parameter': col,
                'Mean': mean,
                'Std': std,
                'Outliers_ZScore': outliers,
                'Outlier_Percent': outlier_pct,
                'Max_ZScore': z_scores.max()
            })

zscore_df = pd.DataFrame(zscore_summary).sort_values('Outlier_Percent', ascending=False)
print("\nZ-Score Outlier Summary (|z| > 3):")
print(zscore_df[['Parameter', 'Outliers_ZScore', 'Outlier_Percent', 'Max_ZScore']].to_string(index=False))

print("\n" + "="*70)
```

```python
# Compare IQR vs Z-Score outlier detection
print("="*70)
print("COMPARISON: IQR vs Z-SCORE OUTLIER DETECTION")
print("="*70)

comparison_data = []

for col in analysis_cols:
    data_clean = df[col].dropna()
    
    if len(data_clean) > 0:
        # IQR method
        Q1 = data_clean.quantile(0.25)
        Q3 = data_clean.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        iqr_outliers = ((data_clean < lower_bound) | (data_clean > upper_bound)).sum()
        
        # Z-score method
        mean = data_clean.mean()
        std = data_clean.std()
        if std > 0:
            z_scores = np.abs((data_clean - mean) / std)
            zscore_outliers = (z_scores > 3).sum()
        else:
            zscore_outliers = 0
        
        comparison_data.append({
            'Parameter': col,
            'IQR_Outliers': iqr_outliers,
            'IQR_Percent': (iqr_outliers / len(data_clean)) * 100,
            'ZScore_Outliers': zscore_outliers,
            'ZScore_Percent': (zscore_outliers / len(data_clean)) * 100,
            'Difference': iqr_outliers - zscore_outliers
        })

comparison_df = pd.DataFrame(comparison_data).sort_values('IQR_Outliers', ascending=False)
print("\nMethod Comparison:")
print(comparison_df.to_string(index=False))

print("\n" + "="*70)
```

```python
# Data quality score per parameter
print("="*70)
print("DATA QUALITY SCORE BY PARAMETER")
print("="*70)

quality_scores = []

for col in analysis_cols:
    data_series = df[col]
    
    # Completeness (% non-null)
    completeness = (1 - data_series.isnull().sum() / len(df)) * 100
    
    # Consistency (1 - outlier percentage using IQR)
    data_clean = data_series.dropna()
    if len(data_clean) > 0:
        Q1 = data_clean.quantile(0.25)
        Q3 = data_clean.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = ((data_clean < lower_bound) | (data_clean > upper_bound)).sum()
        consistency = max(0, (1 - outliers / len(data_clean)) * 100)
    else:
        consistency = 0
    
    # Overall quality score (average of completeness and consistency)
    quality_score = (completeness + consistency) / 2
    
    quality_scores.append({
        'Parameter': col,
        'Completeness': completeness,
        'Consistency': consistency,
        'Quality_Score': quality_score,
        'Grade': 'A' if quality_score >= 95 else 'B' if quality_score >= 85 else 'C' if quality_score >= 75 else 'D'
    })

quality_df = pd.DataFrame(quality_scores).sort_values('Quality_Score', ascending=False)
print("\nQuality Assessment (Score = Average of Completeness & Consistency):")
print(quality_df.to_string(index=False))

print("\n" + "="*70)
print(f"Average Quality Score: {quality_df['Quality_Score'].mean():.2f}")
print(f"Parameters with Grade A (95): {(quality_df['Grade'] == 'A').sum()}")
print(f"Parameters with Grade B (85-94): {(quality_df['Grade'] == 'B').sum()}")
print(f"Parameters with Grade C (75-84): {(quality_df['Grade'] == 'C').sum()}")
print(f"Parameters with Grade D (<75): {(quality_df['Grade'] == 'D').sum()}")
print("="*70)
```

## 2. Stationarity Testing (Critical for Time Series Models)

```python
# Stationarity Tests: Augmented Dickey-Fuller (ADF) and KPSS
from statsmodels.tsa.stattools import adfuller, kpss
import traceback

print("="*80)
print("STATIONARITY TESTING - ADF & KPSS Tests")
print("="*80)
print("\nInterpretation:")
print("  ADF Test - Null Hypothesis: Series has a unit root (non-stationary)")
print("     p-value < 0.05: Reject H0  Series is STATIONARY")
print("     p-value  0.05: Fail to reject H0  Series is NON-STATIONARY")
print("\n  KPSS Test - Null Hypothesis: Series is stationary")
print("     p-value < 0.05: Reject H0  Series is NON-STATIONARY")
print("     p-value  0.05: Fail to reject H0  Series is STATIONARY")
print("="*80)

stationarity_results = []

for param in analysis_cols:
    print(f"\nTesting {param}...")
    ts_data = train_df[param].dropna()
    
    if len(ts_data) > 100:  # Need sufficient data for tests
        try:
            # ADF Test
            print(f"  Running ADF test...")
            adf_result = adfuller(ts_data, autolag='AIC', maxlag=50)
            adf_statistic = adf_result[0]
            adf_pvalue = adf_result[1]
            adf_stationary = "Yes" if adf_pvalue < 0.05 else "No"
            
            # KPSS Test
            print(f"  Running KPSS test...")
            kpss_result = kpss(ts_data, regression='c', nlags='auto')
            kpss_statistic = kpss_result[0]
            kpss_pvalue = kpss_result[1]
            kpss_stationary = "Yes" if kpss_pvalue >= 0.05 else "No"
            
            # Determine final status
            if adf_stationary == "Yes" and kpss_stationary == "Yes":
                status = " STATIONARY"
            elif adf_stationary == "No" and kpss_stationary == "No":
                status = " NON-STATIONARY"
            else:
                status = " INCONCLUSIVE"
            
            stationarity_results.append({
                'Parameter': param,
                'ADF_Statistic': adf_statistic,
                'ADF_pValue': adf_pvalue,
                'ADF_Stationary': adf_stationary,
                'KPSS_Statistic': kpss_statistic,
                'KPSS_pValue': kpss_pvalue,
                'KPSS_Stationary': kpss_stationary,
                'Status': status
            })
            print(f"   Complete: {status}")
        except Exception as e:
            print(f"   Error: {str(e)}")
            traceback.print_exc()

stationarity_df = pd.DataFrame(stationarity_results)

print("\n" + "="*80)
print("STATIONARITY TEST RESULTS")
print("="*80)
if len(stationarity_df) > 0:
    print(stationarity_df[['Parameter', 'ADF_pValue', 'ADF_Stationary', 
                            'KPSS_pValue', 'KPSS_Stationary', 'Status']].to_string(index=False))

    # Summary statistics
    print("\n" + "="*80)
    print("SUMMARY:")
    print(f"   Stationary series: {(stationarity_df['Status'] == ' STATIONARY').sum()}")
    print(f"   Non-stationary series: {(stationarity_df['Status'] == ' NON-STATIONARY').sum()}")
    print(f"   Inconclusive: {(stationarity_df['Status'] == ' INCONCLUSIVE').sum()}")
    print("\n TIP: Non-stationary series may need differencing before modeling")
else:
    print("No results - all tests failed. Check errors above.")
print("="*80)
```

## 4. Spectral Analysis and Periodicity Detection

```python
# Fourier Transform Analysis - Detect dominant frequencies and periodicities
from scipy import signal
from scipy.fft import fft, fftfreq

print("="*70)
print("SPECTRAL ANALYSIS - PERIODICITY DETECTION")
print("="*70)

key_params = ['Dst_Index_nT', 'Flow_Speed_km_s', 'IMF_Mag_Avg_nT', 'Bz_GSM_nT']
fig, axes = plt.subplots(len(key_params), 2, figsize=(18, 4*len(key_params)))

periodicity_results = []

for idx, param in enumerate(key_params):
    ts_data = train_df[param].dropna().values
    
    # Perform FFT
    n = len(ts_data)
    yf = fft(ts_data)
    xf = fftfreq(n, d=1.0)[:n//2]  # Frequency bins (cycles per hour)
    power = 2.0/n * np.abs(yf[0:n//2])
    
    # Time domain plot
    axes[idx, 0].plot(range(len(ts_data)), ts_data, linewidth=0.5, color='blue')
    axes[idx, 0].set_title(f'Time Domain - {param}', fontsize=10)
    axes[idx, 0].set_xlabel('Time (hours)', fontsize=9)
    axes[idx, 0].set_ylabel('Value', fontsize=9)
    axes[idx, 0].grid(True, alpha=0.3)
    
    # Frequency domain plot (power spectrum)
    axes[idx, 1].plot(xf, power, linewidth=0.5, color='red')
    axes[idx, 1].set_title(f'Power Spectrum - {param}', fontsize=10)
    axes[idx, 1].set_xlabel('Frequency (cycles/hour)', fontsize=9)
    axes[idx, 1].set_ylabel('Power', fontsize=9)
    axes[idx, 1].set_xlim(0, 0.1)  # Focus on low frequencies
    axes[idx, 1].grid(True, alpha=0.3)
    
    # Find dominant frequencies
    top_freq_idx = np.argsort(power)[-5:][::-1]  # Top 5 frequencies
    for freq_idx in top_freq_idx:
        if xf[freq_idx] > 0:  # Exclude DC component
            period_hours = 1 / xf[freq_idx]
            periodicity_results.append({
                'Parameter': param,
                'Frequency': xf[freq_idx],
                'Period_Hours': period_hours,
                'Period_Days': period_hours / 24,
                'Power': power[freq_idx]
            })

plt.suptitle('Spectral Analysis - Frequency Domain', fontsize=14, y=0.999)
plt.tight_layout()
plt.show()

# Display periodicity results
periodicity_df = pd.DataFrame(periodicity_results).sort_values('Power', ascending=False)
print("\n" + "="*70)
print("DOMINANT PERIODICITIES DETECTED:")
print("="*70)
print(periodicity_df.head(10).to_string(index=False))

print("\n INSIGHTS:")
print("   Periodicities help identify recurring patterns (daily, weekly, etc.)")
print("   Use this information for feature engineering and model design")
print("   Common space weather cycles: 27-day solar rotation, ~11-year solar cycle")
print("="*70)
```

## 5. Advanced Lag Feature Engineering for Deep Learning

```python
# Create lag features and rolling statistics for enhanced model performance
print("="*70)
print("ADVANCED LAG FEATURE ENGINEERING")
print("="*70)

# Define lag windows
lag_hours = [1, 3, 6, 12, 24, 48, 72, 168]  # 1h, 3h, 6h, 12h, 1d, 2d, 3d, 7d
rolling_windows = [6, 12, 24, 48, 72]  # 6h, 12h, 1d, 2d, 3d

# Select key parameters for lag features
lag_params = ['Dst_Index_nT', 'Flow_Speed_km_s', 'IMF_Mag_Avg_nT', 'Bz_GSM_nT']

# Create a copy of scaled data for feature engineering
df_lag = train_scaled.copy()

print(f"\nCreating lag features for {len(lag_params)} parameters...")
print(f"Lag windows: {lag_hours}")
print(f"Rolling windows: {rolling_windows}")

feature_count_before = len(df_lag.columns)

# Create lag features
for param in lag_params:
    for lag in lag_hours:
        df_lag[f'{param}_lag_{lag}h'] = df_lag[param].shift(lag)

# Create rolling statistics (mean, std, min, max)
for param in lag_params:
    for window in rolling_windows:
        df_lag[f'{param}_rolling_mean_{window}h'] = df_lag[param].rolling(window=window).mean()
        df_lag[f'{param}_rolling_std_{window}h'] = df_lag[param].rolling(window=window).std()
        df_lag[f'{param}_rolling_min_{window}h'] = df_lag[param].rolling(window=window).min()
        df_lag[f'{param}_rolling_max_{window}h'] = df_lag[param].rolling(window=window).max()

# Create rate of change features
for param in lag_params:
    df_lag[f'{param}_diff_1h'] = df_lag[param].diff(1)
    df_lag[f'{param}_diff_24h'] = df_lag[param].diff(24)
    df_lag[f'{param}_pct_change_1h'] = df_lag[param].pct_change(1)
    df_lag[f'{param}_pct_change_24h'] = df_lag[param].pct_change(24)

feature_count_after = len(df_lag.columns)
new_features = feature_count_after - feature_count_before

print(f"\n Feature engineering complete!")
print(f"   Original features: {feature_count_before}")
print(f"   New lag features created: {new_features}")
print(f"   Total features: {feature_count_after}")

# Remove rows with NaN due to lagging/rolling (keep data integrity)
print(f"\nRows before cleaning: {len(df_lag):,}")
df_lag_clean = df_lag.dropna()
print(f"Rows after cleaning: {len(df_lag_clean):,}")
print(f"Rows removed: {len(df_lag) - len(df_lag_clean):,}")

print("\n" + "="*70)
print("SAMPLE OF ENGINEERED FEATURES:")
print("="*70)
lag_feature_cols = [col for col in df_lag_clean.columns if 'lag_' in col or 'rolling_' in col or 'diff_' in col or 'pct_change_' in col]
print(f"\nTotal engineered features: {len(lag_feature_cols)}")
print(f"\nExample features:")
for col in lag_feature_cols[:10]:
    print(f"   {col}")
print(f"  ... and {len(lag_feature_cols) - 10} more")

print("\n BENEFIT: Lag features capture temporal dependencies explicitly")
print("   Useful for models like Random Forest, XGBoost, and Neural Networks")
print("="*70)
```

```python
# Alternative: Create sequences with smaller step size to fit in memory
# Or use memory-mapped files to save directly to disk

import tempfile
import os
import gc

def create_sequences_to_disk(data, target_col, feature_cols, lookback, forecast_horizon, step,
                              output_prefix):
    """
    Create sequences and save directly to .npy files (batch-by-batch to avoid memory issues)
    """
    n_samples = len(data)
    n_sequences = (n_samples - lookback - forecast_horizon) // step + 1
    n_features = len(feature_cols)
    
    print(f"  Total sequences to create: {n_sequences:,}")
    print(f"  Estimated memory needed: {(n_sequences * lookback * n_features * 4) / (1024**3):.2f} GB for X")
    print(f"  Estimated memory needed: {(n_sequences * forecast_horizon * 4) / (1024**3):.2f} GB for y")
    
    # Create memory-mapped files (disk-based arrays)
    X_file = f'{output_prefix}_X_temp.dat'
    y_file = f'{output_prefix}_y_temp.dat'
    
    print(f"  Creating disk-based arrays...")
    X_memmap = np.memmap(X_file, dtype='float32', mode='w+', 
                         shape=(n_sequences, lookback, n_features))
    y_memmap = np.memmap(y_file, dtype='float32', mode='w+',
                         shape=(n_sequences, forecast_horizon))
    
    # Get data as numpy arrays
    X_data = data[feature_cols].values.astype(np.float32)
    y_data = data[target_col].values.astype(np.float32)
    
    print(f"  Writing sequences to disk (this may take a few minutes)...")
    seq_idx = 0
    for i in range(0, n_samples - lookback - forecast_horizon + 1, step):
        X_memmap[seq_idx] = X_data[i:i+lookback]
        y_memmap[seq_idx] = y_data[i+lookback:i+lookback+forecast_horizon]
        seq_idx += 1
        
        if seq_idx % 20000 == 0:
            print(f"    Progress: {seq_idx:,}/{n_sequences:,} ({seq_idx/n_sequences*100:.1f}%)")
            X_memmap.flush()
            y_memmap.flush()
    
    # Final flush
    X_memmap.flush()
    y_memmap.flush()
    
    print(f"   Created {seq_idx:,} sequences")
    
    # Convert memmap to regular .npy files
    print(f"  Converting to .npy format...")
    np.save(f'{output_prefix}_X.npy', X_memmap)
    np.save(f'{output_prefix}_y.npy', y_memmap)
    
    shapes = (X_memmap.shape, y_memmap.shape)
    
    # Cleanup
    del X_memmap, y_memmap
    gc.collect()
    
    # Remove temp files
    try:
        os.remove(X_file)
        os.remove(y_file)
    except:
        pass
    
    return shapes

print(" Disk-based sequence creation function loaded")
```

```python
# Prepare sequences for LSTM/GRU/Transformer models (Memory-Efficient Version)
print("="*70)
print("SEQUENCE PREPARATION FOR DEEP LEARNING MODELS")
print("="*70)

def create_sequences_efficient(data, target_col, feature_cols, lookback=72, forecast_horizon=168, step=1):
    """
    Create sequences for time series forecasting - Memory efficient version.
    
    Args:
        data: DataFrame with features
        target_col: Column name to predict
        feature_cols: List of feature column names to use
        lookback: Number of past time steps to use as input (hours)
        forecast_horizon: Number of future steps to predict (hours, 168h = 7 days)
        step: Step size between sequences (1=all sequences, higher=fewer sequences)
    
    Returns:
        X: Input sequences (samples, lookback, features)
        y: Target values (samples, forecast_horizon)
    """
    # Pre-allocate arrays for better memory efficiency
    n_sequences = len(range(lookback, len(data) - forecast_horizon + 1, step))
    n_features = len(feature_cols)
    
    print(f"  Allocating memory for {n_sequences:,} sequences...")
    X = np.zeros((n_sequences, lookback, n_features), dtype=np.float32)
    y = np.zeros((n_sequences, forecast_horizon), dtype=np.float32)
    
    # Convert to numpy arrays once for speed
    data_values = data[feature_cols].values.astype(np.float32)
    target_values = data[target_col].values.astype(np.float32)
    
    # Create sequences using numpy slicing
    seq_idx = 0
    for i in range(lookback, len(data) - forecast_horizon + 1, step):
        X[seq_idx] = data_values[i-lookback:i]
        y[seq_idx] = target_values[i:i+forecast_horizon]
        seq_idx += 1
        
        if seq_idx % 10000 == 0:
            print(f"  Progress: {seq_idx:,} / {n_sequences:,} sequences created")
    
    return X, y

# Define parameters
LOOKBACK_WINDOW = 72  # 3 days of history
FORECAST_HORIZON = 168  # 7 days forecast
TARGET_PARAMETERS = ['Dst_Index_nT', 'ap_index_nT', 'Sunspot_Number', 'Kp_10']  # All 4 targets
SEQUENCE_STEP = 1  # Use ALL data points (step=1 means every hour)

# Select key features (reduce from 155 to essential features only)
KEY_FEATURES = [
    'IMF_Mag_Avg_nT', 'IMF_Lat_deg', 'IMF_Long_deg', 'Bz_GSM_nT',
    'Proton_Density_n_cc', 'Flow_Speed_km_s', 'Flow_Longitude_deg', 'Flow_Latitude_deg',
    'Alpha_Proton_Ratio', 'Kp_10', 'Sunspot_Number', 'Dst_Index_nT', 'ap_index_nT', 'F10.7_Index',
    'Hour_Sin', 'Hour_Cos', 'DOY_Sin', 'DOY_Cos'
]

print(f"\nSequence Configuration:")
print(f"   Lookback window: {LOOKBACK_WINDOW} hours ({LOOKBACK_WINDOW/24:.1f} days)")
print(f"   Forecast horizon: {FORECAST_HORIZON} hours ({FORECAST_HORIZON/24:.1f} days)")
print(f"   Target parameters: {TARGET_PARAMETERS}")
print(f"   Features used: {len(KEY_FEATURES)} (reduced from 155 for memory efficiency)")
print(f"   Sequence step: {SEQUENCE_STEP} hours (using full data)")

# Create sequences for all 4 target parameters
# Use DISK-BASED approach to avoid RAM limitations
import gc

summary_info = []

for target_param in TARGET_PARAMETERS:
    print(f"\n{'='*70}")
    print(f"Creating sequences for: {target_param}")
    print(f"{'='*70}")
    
    target_clean = target_param.replace('_', '').replace('.', '')
    
    # Create sequences from training data (saved directly to disk)
    print(f"\nTraining data...")
    train_shapes = create_sequences_to_disk(
        train_scaled, 
        target_col=target_param,
        feature_cols=KEY_FEATURES,
        lookback=LOOKBACK_WINDOW,
        forecast_horizon=FORECAST_HORIZON,
        step=SEQUENCE_STEP,
        output_prefix=f'../train_{target_clean}'
    )
    
    # Create sequences from test data (saved directly to disk)
    print(f"\nTest data...")
    test_shapes = create_sequences_to_disk(
        test_scaled,
        target_col=target_param,
        feature_cols=KEY_FEATURES,
        lookback=LOOKBACK_WINDOW,
        forecast_horizon=FORECAST_HORIZON,
        step=SEQUENCE_STEP,
        output_prefix=f'../test_{target_clean}'
    )
    
    print(f"\n {target_param} sequences created:")
    print(f"  Train: X={train_shapes[0]}, y={train_shapes[1]}")
    print(f"  Test:  X={test_shapes[0]}, y={test_shapes[1]}")
    print(f"  Files: train_{target_clean}_X.npy, train_{target_clean}_y.npy")
    print(f"  Files: test_{target_clean}_X.npy, test_{target_clean}_y.npy")
    
    # Store summary info
    summary_info.append({
        'target': target_param,
        'train_sequences': train_shapes[0][0],
        'test_sequences': test_shapes[0][0],
        'input_shape': (train_shapes[0][1], train_shapes[0][2]),
        'output_steps': train_shapes[1][1]
    })
    
    # Force garbage collection
    gc.collect()
    print(f"   Memory cleaned")

print("\n" + "="*70)
print("FINAL SUMMARY - ALL TARGET PARAMETERS")
print("="*70)

for info in summary_info:
    print(f"\n{info['target']}:")
    print(f"  Training: {info['train_sequences']:,} sequences")
    print(f"  Test:     {info['test_sequences']:,} sequences")
    print(f"  Input:    {info['input_shape'][0]} timesteps  {info['input_shape'][1]} features")
    print(f"  Output:   {info['output_steps']} forecast steps")

print("\n" + "="*70)
print(" USAGE FOR DEEP LEARNING:")
print("="*70)
print("""
# Example: Load and train model for Dst_Index_nT
X_train = np.load('../X_train_DstIndexnT.npy')
y_train = np.load('../y_train_DstIndexnT.npy')
X_test = np.load('../X_test_DstIndexnT.npy')
y_test = np.load('../y_test_DstIndexnT.npy')

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(64, return_sequences=False),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dense(y_train.shape[1])  # Output: 168 values (7-day forecast)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
""")
print("="*70)

print("\n All sequences created and saved successfully!")
print(f" Files saved: X_train_*.npy, y_train_*.npy, X_test_*.npy, y_test_*.npy")
print("="*70)
```

## Model 1: Baseline LSTM

Start with a simple LSTM architecture for all 4 targets

```python
def build_lstm_baseline(input_shape, output_steps, name='LSTM_Baseline'):
    """
    Baseline LSTM model for time series forecasting
    
    Architecture:
    - LSTM layer with 128 units
    - Dropout for regularization
    - LSTM layer with 64 units
    - Dense layers for output
    """
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=input_shape, name='lstm_1'),
        Dropout(0.2, name='dropout_1'),
        LSTM(64, return_sequences=False, name='lstm_2'),
        Dropout(0.2, name='dropout_2'),
        Dense(64, activation='relu', name='dense_1'),
        Dense(output_steps, name='output')
    ], name=name)
    
    return model

# Model configuration
INPUT_SHAPE = (72, 18)  # (lookback_window, n_features)
OUTPUT_STEPS = 168      # 7 days forecast

print("="*70)
print("BASELINE LSTM MODEL ARCHITECTURE")
print("="*70)

# Build model
lstm_baseline = build_lstm_baseline(INPUT_SHAPE, OUTPUT_STEPS)

# Compile model
lstm_baseline.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae', 'mse']
)

# Display architecture
lstm_baseline.summary()

print("\n Baseline LSTM model created")
print(f"  Total parameters: {lstm_baseline.count_params():,}")
```

```python
# Train baseline LSTM for first target (Dst_Index_nT)
print("="*70)
print("TRAINING BASELINE LSTM - Dst_Index_nT")
print("="*70)

# Load sequences for Dst_Index_nT
X_train = np.load('../train_DstIndexnT_X.npy', mmap_mode='r')
y_train = np.load('../train_DstIndexnT_y.npy', mmap_mode='r')
X_test = np.load('../test_DstIndexnT_X.npy', mmap_mode='r')
y_test = np.load('../test_DstIndexnT_y.npy', mmap_mode='r')

print(f"\nData loaded:")
print(f"  X_train: {X_train.shape}")
print(f"  y_train: {y_train.shape}")
print(f"  X_test: {X_test.shape}")
print(f"  y_test: {y_test.shape}")

# Define callbacks
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    ),
    ModelCheckpoint(
        '../models/lstm_baseline_dst.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

print("\n Starting training...")
print(f"  Batch size: 32")
print(f"  Epochs: 50 (with early stopping)")
print(f"  Validation split: 20%")

# Train model
history = lstm_baseline.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

print("\n Training completed!")

# Evaluate on test set
test_loss, test_mae, test_mse = lstm_baseline.evaluate(X_test, y_test, verbose=0)
print(f"\n Test Set Performance:")
print(f"  MSE: {test_mse:.4f}")
print(f"  MAE: {test_mae:.4f}")
print(f"  RMSE: {np.sqrt(test_mse):.4f}")
```

```python
# Visualize training history
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss plot
axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
axes[0].set_title('Model Loss - Dst_Index_nT', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('MSE Loss', fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# MAE plot
axes[1].plot(history.history['mae'], label='Training MAE', linewidth=2)
axes[1].plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
axes[1].set_title('Model MAE - Dst_Index_nT', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('MAE', fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f" Best validation loss: {min(history.history['val_loss']):.4f}")
print(f" Best validation MAE: {min(history.history['val_mae']):.4f}")
```

## Model 2: Bidirectional LSTM

Enhanced model with bidirectional processing for better temporal pattern capture

```python
def build_bidirectional_lstm(input_shape, output_steps, name='BiLSTM'):
    """
    Bidirectional LSTM model for improved temporal pattern recognition
    
    Architecture:
    - Bidirectional LSTM (128 units forward + 128 backward)
    - Dropout
    - Bidirectional LSTM (64 units forward + 64 backward)
    - Dense layers for output
    """
    model = Sequential([
        Bidirectional(LSTM(128, return_sequences=True), input_shape=input_shape, name='bilstm_1'),
        Dropout(0.3, name='dropout_1'),
        Bidirectional(LSTM(64, return_sequences=False), name='bilstm_2'),
        Dropout(0.3, name='dropout_2'),
        Dense(128, activation='relu', name='dense_1'),
        Dropout(0.2, name='dropout_3'),
        Dense(64, activation='relu', name='dense_2'),
        Dense(output_steps, name='output')
    ], name=name)
    
    return model

print("="*70)
print("BIDIRECTIONAL LSTM MODEL ARCHITECTURE")
print("="*70)

# Build model
bilstm_model = build_bidirectional_lstm(INPUT_SHAPE, OUTPUT_STEPS)

# Compile
bilstm_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae', 'mse']
)

# Display architecture
bilstm_model.summary()

print("\n Bidirectional LSTM model created")
print(f"  Total parameters: {bilstm_model.count_params():,}")
print(f"  Improvement: Processes sequences forward AND backward")
```

## Model 3: GRU (Gated Recurrent Unit)

Faster alternative to LSTM with comparable performance

```python
def build_gru_model(input_shape, output_steps, name='GRU_Model'):
    """
    GRU model - faster training than LSTM with similar performance
    
    Architecture:
    - GRU layer (128 units)
    - Dropout
    - GRU layer (64 units)
    - Dense layers for output
    """
    model = Sequential([
        GRU(128, return_sequences=True, input_shape=input_shape, name='gru_1'),
        Dropout(0.2, name='dropout_1'),
        GRU(64, return_sequences=False, name='gru_2'),
        Dropout(0.2, name='dropout_2'),
        Dense(64, activation='relu', name='dense_1'),
        Dense(output_steps, name='output')
    ], name=name)
    
    return model

print("="*70)
print("GRU MODEL ARCHITECTURE")
print("="*70)

# Build model
gru_model = build_gru_model(INPUT_SHAPE, OUTPUT_STEPS)

# Compile
gru_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae', 'mse']
)

# Display architecture
gru_model.summary()

print("\n GRU model created")
print(f"  Total parameters: {gru_model.count_params():,}")
print(f"  Advantage: Faster training, fewer parameters than LSTM")
```

## Model 4: CNN-LSTM Hybrid

Combines CNN for feature extraction with LSTM for temporal modeling

```python
def build_cnn_lstm_model(input_shape, output_steps, name='CNN_LSTM'):
    """
    CNN-LSTM hybrid model
    
    Architecture:
    - Conv1D layers for local feature extraction
    - LSTM layers for temporal dependencies
    - Dense layers for output
    """
    model = Sequential([
        Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape, name='conv1d_1'),
        Conv1D(filters=64, kernel_size=3, activation='relu', name='conv1d_2'),
        MaxPooling1D(pool_size=2, name='maxpool'),
        LSTM(128, return_sequences=True, name='lstm_1'),
        Dropout(0.2, name='dropout_1'),
        LSTM(64, return_sequences=False, name='lstm_2'),
        Dropout(0.2, name='dropout_2'),
        Dense(128, activation='relu', name='dense_1'),
        Dense(64, activation='relu', name='dense_2'),
        Dense(output_steps, name='output')
    ], name=name)
    
    return model

print("="*70)
print("CNN-LSTM HYBRID MODEL ARCHITECTURE")
print("="*70)

# Build model
cnn_lstm_model = build_cnn_lstm_model(INPUT_SHAPE, OUTPUT_STEPS)

# Compile
cnn_lstm_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae', 'mse']
)

# Display architecture
cnn_lstm_model.summary()

print("\n CNN-LSTM model created")
print(f"  Total parameters: {cnn_lstm_model.count_params():,}")
print(f"  Advantage: CNN extracts local patterns, LSTM captures long-term dependencies")
```

## Training Function for Multiple Targets

Create a reusable function to train any model on all 4 targets

```python
def train_model_for_all_targets(model_builder_func, model_name, epochs=50, batch_size=32):
    """
    Train a model for all 4 target parameters
    
    Args:
        model_builder_func: Function that builds the model
        model_name: Name for saving models
        epochs: Number of training epochs
        batch_size: Batch size for training
        
    Returns:
        results: Dictionary with training results for each target
    """
    target_files = {
        'Dst_Index_nT': 'DstIndexnT',
        'ap_index_nT': 'apindexnT',
        'Sunspot_Number': 'SunspotNumber',
        'Kp_10': 'Kp10'
    }
    
    results = {}
    
    for target_name, file_suffix in target_files.items():
        print("\n" + "="*70)
        print(f"TRAINING {model_name} - {target_name}")
        print("="*70)
        
        # Load data
        X_train = np.load(f'../train_{file_suffix}_X.npy', mmap_mode='r')
        y_train = np.load(f'../train_{file_suffix}_y.npy', mmap_mode='r')
        X_test = np.load(f'../test_{file_suffix}_X.npy', mmap_mode='r')
        y_test = np.load(f'../test_{file_suffix}_y.npy', mmap_mode='r')
        
        print(f"\n Data loaded: {X_train.shape[0]:,} train, {X_test.shape[0]:,} test")
        
        # Build model
        model = model_builder_func(INPUT_SHAPE, OUTPUT_STEPS, name=f'{model_name}_{target_name}')
        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='mse',
            metrics=['mae', 'mse']
        )
        
        # Callbacks
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
        ]
        
        # Train
        print(f"\n Training...")
        history = model.fit(
            X_train, y_train,
            validation_split=0.2,
            epochs=epochs,
            batch_size=batch_size,
            callbacks=callbacks,
            verbose=0
        )
        
        # Evaluate
        test_loss, test_mae, test_mse = model.evaluate(X_test, y_test, verbose=0)
        
        # Save model
        model.save(f'../models/{model_name}_{file_suffix}.keras')
        
        # Store results
        results[target_name] = {
            'history': history,
            'test_loss': test_loss,
            'test_mae': test_mae,
            'test_mse': test_mse,
            'test_rmse': np.sqrt(test_mse),
            'best_val_loss': min(history.history['val_loss']),
            'epochs_trained': len(history.history['loss'])
        }
        
        print(f"\n Training completed!")
        print(f"  Epochs: {results[target_name]['epochs_trained']}")
        print(f"  Best val loss: {results[target_name]['best_val_loss']:.4f}")
        print(f"  Test MAE: {test_mae:.4f}")
        print(f"  Test RMSE: {results[target_name]['test_rmse']:.4f}")
        
        # Clean up memory
        del X_train, y_train, X_test, y_test, model
        import gc
        gc.collect()
    
    return results

print(" Training function created")
print("  Ready to train models for all 4 targets")
```

## Model Comparison Summary

Create a summary comparing all models across all targets

```python
def create_model_comparison_report(all_results):
    """
    Create comprehensive comparison report of all models
    
    Args:
        all_results: Dictionary with results from all models
    """
    print("="*80)
    print("COMPREHENSIVE MODEL COMPARISON REPORT")
    print("="*80)
    
    # Prepare data for comparison
    comparison_data = []
    
    for model_name, targets_results in all_results.items():
        for target_name, metrics in targets_results.items():
            comparison_data.append({
                'Model': model_name,
                'Target': target_name,
                'Test MAE': metrics['test_mae'],
                'Test RMSE': metrics['test_rmse'],
                'Best Val Loss': metrics['best_val_loss'],
                'Epochs': metrics['epochs_trained']
            })
    
    comparison_df = pd.DataFrame(comparison_data)
    
    # Display full comparison table
    print("\n DETAILED RESULTS:")
    print(comparison_df.to_string(index=False))
    
    # Find best models for each target
    print("\n" + "="*80)
    print(" BEST MODELS BY TARGET")
    print("="*80)
    
    for target in comparison_df['Target'].unique():
        target_df = comparison_df[comparison_df['Target'] == target]
        best_model = target_df.loc[target_df['Test MAE'].idxmin()]
        print(f"\n{target}:")
        print(f"  Best Model: {best_model['Model']}")
        print(f"  MAE: {best_model['Test MAE']:.4f}")
        print(f"  RMSE: {best_model['Test RMSE']:.4f}")
    
    # Overall best model
    print("\n" + "="*80)
    print(" OVERALL CHAMPION")
    print("="*80)
    
    avg_performance = comparison_df.groupby('Model').agg({
        'Test MAE': 'mean',
        'Test RMSE': 'mean'
    }).round(4)
    
    best_overall = avg_performance['Test MAE'].idxmin()
    print(f"\nBest Overall Model: {best_overall}")
    print(f"\nAverage Performance Across All Targets:")
    print(avg_performance.to_string())
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Model Performance Comparison', fontsize=16, fontweight='bold')
    
    targets = comparison_df['Target'].unique()
    
    for idx, target in enumerate(targets):
        row, col = idx // 2, idx % 2
        target_df = comparison_df[comparison_df['Target'] == target]
        
        x = np.arange(len(target_df))
        width = 0.35
        
        axes[row, col].bar(x - width/2, target_df['Test MAE'], width, label='MAE', alpha=0.8)
        axes[row, col].bar(x + width/2, target_df['Test RMSE'], width, label='RMSE', alpha=0.8)
        axes[row, col].set_title(target, fontsize=12, fontweight='bold')
        axes[row, col].set_xlabel('Model', fontsize=10)
        axes[row, col].set_ylabel('Error', fontsize=10)
        axes[row, col].set_xticks(x)
        axes[row, col].set_xticklabels(target_df['Model'], rotation=45, ha='right')
        axes[row, col].legend()
        axes[row, col].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return comparison_df

print(" Comparison report function created")
```
