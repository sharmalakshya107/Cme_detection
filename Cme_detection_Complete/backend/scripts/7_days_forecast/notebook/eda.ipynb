```python
# Check all positions of null values for each column
for col in data.columns:
    null_positions = data[data[col].isna()].index.tolist()
    null_count = len(null_positions)
    if null_count > 0:
        print(f"\n{col}: {null_count} null values")
        print(f"  All positions: {null_positions}")
```

```python
# Find columns that share the same null positions
import pandas as pd

# Get null positions for each column
null_patterns = {}
for col in data.columns:
    null_positions = tuple(data[data[col].isna()].index.tolist())
    if null_positions:  # Only include columns with nulls
        if null_positions not in null_patterns:
            null_patterns[null_positions] = []
        null_patterns[null_positions].append(col)

# Display groups of columns with identical null patterns
print("Columns grouped by identical null positions:\n")
for i, (positions, columns) in enumerate(null_patterns.items(), 1):
    if len(columns) > 1:  # Only show groups with multiple columns
        print(f"Group {i}: {len(columns)} columns with {len(positions)} null values at same positions")
        print(f"  Columns: {columns}")
        print(f"  All positions: {list(positions)}")
        print()

# Summary
print("\nSummary:")
print(f"Total unique null patterns: {len(null_patterns)}")
groups_with_multiple = sum(1 for cols in null_patterns.values() if len(cols) > 1)
print(f"Groups with multiple columns sharing same nulls: {groups_with_multiple}")
```

```python
# Remove rows at specific positions that have all null values
positions_to_remove = [8808, 8809, 13226, 13227, 17644, 17645, 22038, 22039, 26456, 26457, 30850, 30851, 35268, 35269, 39662, 39663, 44080, 44081, 48498, 48499, 52916, 52917, 57310, 57311, 61728, 61729, 66122, 66123, 70540, 70541, 74934, 74935, 79352, 79353, 83770, 83771, 88188, 88189, 92582, 92583]

print(f"Original dataset: {len(data)} rows")
print(f"Positions to remove: {len(positions_to_remove)} rows")

# Drop rows at these specific positions
data_cleaned = data.drop(index=positions_to_remove, errors='ignore')

rows_removed = len(data) - len(data_cleaned)
print(f"Rows actually removed: {rows_removed}")
print(f"Cleaned dataset: {len(data_cleaned)} rows")

# Update the data variable and reset index
data = data_cleaned.reset_index(drop=True)
print(f"\n Data updated with reset index: {len(data)} rows remaining")
```

```python
# Create heatmap for file2
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Select only numeric columns for correlation
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()

# Calculate correlation matrix
correlation_matrix = data[numeric_cols].corr()

# Create heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm', 
            center=0,
            square=True,
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})

plt.title('Correlation Heatmap', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```
